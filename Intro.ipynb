## Shapefile
- The major file format for geographic data
- usually downloaded as zip file

1. shp - shapefile - actual geometry of file
2. dbf - database file - additional info outside of the shape
3. prj - projection file - gives all the assumptions about earth we are making like area
4. shx - index file - basically makes shp file a little bit faster
